```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
for page in range(1,51):
    
    website = "https://www.carfax.com/Used-Cars-in-Silver-Spring-MD_c23683".format(page)
    response = requests.get(website)
    # soup object (to get html element)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Results List
    cars = soup.find('div', class_="srp-list-container srp-list-container--srp").find_all('article')

    for car in cars:
        model = car.find('div', class_='listing-header').h4.text
        price = car.find('span', class_ = 'srp-list-item-price').text
    #     mileage = car.find('span', class_ = 'srp-list-item-basic-info-value').text
        engine = car.find('div', class_ = "srp-list-item-basic-info srp-list-item-special-features").text

        print(model, price, engine)
    
```
